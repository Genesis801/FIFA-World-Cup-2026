# Analysis 1
## Did Football Become More Defensive?

## Business Question

#### People often say:

Modern football is more tactical and defensive.

### Can we prove that statistically?

```python
# Checking Goals per tournament:
tournaments[['wc_year','total_goals']]
```

```python
# Visualizing Goals per tournament:
plt.figure(figsize=(14,6))

sns.lineplot(
    data=tournaments,
    x='wc_year',
    y='total_goals',
    marker='o'
)

plt.title("Total Goals by World Cup")
plt.xlabel("Year")
plt.ylabel("Goals")
plt.show()
```

```python
# Normalize by Matches. Why?
# Early tournaments had fewer teams. Comparing raw goals is misleading.
tournaments['goals_per_match'] = (
    tournaments['total_goals']
    /
    tournaments['total_matches']
)
```

```python
sns.lineplot(
    data=tournaments,
    x='wc_year',
    y='goals_per_match',
    marker='o'
)

plt.title("Goals per Match Through History")
plt.show()
```

#### Comparing Eras
Comparing eras is more meaningful than individual years.

```python
era_goals = (
    tournaments
    .groupby("format_era")
    ["goals_per_match"]
    .mean()
    .sort_values()
)

era_goals
```

```python
era_goals.plot(
    kind="bar"
)

plt.title("Average Goals per Match by Era")
plt.ylabel("Goals")
plt.show()
```

# Statistical Test
## Why?

### Visualization alone isn't enough.

### Test:

H0:
No difference between eras

H1:
Goals per match differs

```python
groups = []

for era in tournaments['format_era'].unique():
    groups.append(
        tournaments[
            tournaments['format_era']==era
        ]['goals_per_match']
    )

print("p-value:", stats.f_oneway(*groups).pvalue)

if stats.f_oneway(*groups).pvalue < 0.05:
    print("Goals per match vary significantly between eras.")
else:
    print("No significant differences between eras.")
```

## Conclusion
#### 1. Average goals per match declined from 3.8 in early tournaments to 2.5 in modern tournaments.This suggests football has become more structured and defensive.
#### 2. Group Stage football is more defensive and tactical than K-O Stages with higher stakes.

# Analysis 2
## Which Confederation Dominates World Football?
### Business Question

Is UEFA really dominant?

Or has CONMEBOL been equally successful?

```python
confed_lookup = {
    "Brazil":"CONMEBOL",
    "Argentina":"CONMEBOL",
    "Uruguay":"CONMEBOL",
    "West Germany":"UEFA",
    "Germany":"UEFA",
    "Italy":"UEFA",
    "France":"UEFA",
    "England":"UEFA",
    "Spain":"UEFA",
    "Nigeria":"CAF",
    "Japan":"AFC",
    "USA":"CONCACAF"
}

# convert to DataFrame 
confed_df = pd.DataFrame(list(confed_lookup.items()), columns=['team', 'confederation'])
```

```python
teams = teams.merge(
    confed_df,
    on='team',
    how='left'
)
```

```python
teams.tail()
```

```python
# Join champion with confederation.
champions = tournaments[
    ['wc_year','champion']
]

champions = champions.merge(
    confed_df,
    left_on='champion',
    right_on='team',
    how='left'
)

champions.head()
```

```python
champions['confederation'].value_counts()
```

```python
champions['confederation']\
    .value_counts()\
    .plot(kind='bar')

plt.title("World Cup Titles by Confederation")
plt.show()
```

### Confederation Win Rate

Create match-level analysis.

```python
# Calculate winner confederation.
winner_confed = matches.merge(
    confed_df,
    left_on='winning_team',
    right_on='team',
    how='left'
)

winner_confed['confederation'].value_counts()
```

```python
sns.countplot(
    data=winner_confed,
    x='confederation'
)

plt.title("Match Wins by Confederation")
plt.show()
```

```python
# Confederation dominance over time.
champions.groupby(
    ['wc_year','confederation']
).size().unstack(fill_value=0)
```

```python
# Heatmap of confederation wins over time.
heatmap_df = champions.groupby(
    ['wc_year','confederation']
).size().unstack(fill_value=0)

sns.heatmap(
    heatmap_df.T,
    cmap="Blues"
)
```

# Conclusion
1. UEFA dominates after 2000.

2. CONMEBOL dominated earlier decades.

3. CAF, AFC, CONCACAF have never won.

# Analysis 3
## Home Advantage
### Business Question

Do hosts perform better than expected?

This is a fantastic sports analytics question.

```python
hosts = tournaments[
    ['wc_year','host_nation']
]
host_teams = teams.merge(
    hosts,
    on='wc_year',
    how='left'
)

# host_teams.head()
host_teams['is_host'] = (
    host_teams['team']
    ==
    host_teams['host_nation_y']
)
```

```python
# Converting stages into scores
stage_score = {
    "Group Stage":1,
    "Round of 16":2,
    "Quarter-finals":3,
    "Semi-finals":4,
    "Runner-up":5,
    "Champion":6
}
host_teams['performance_score'] = (
    host_teams['final_stage_reached']
    .map(stage_score)
)
```

```python
# Calculating expected performance
host_teams['elo_percentile'] = (
    host_teams['elo_rating_approx']
    .rank(pct=True)
)

# Comapre host vs non-host performance
host_teams.groupby(
    'is_host'
)['performance_score'].mean()
```

```python
sns.boxplot(
    data=host_teams,
    x='is_host',
    y='performance_score'
)

plt.title("Host Performance")
plt.show()
```

```python
# Host Outperformance metric

# Expected score
host_teams['expected_score'] = (
    host_teams['elo_percentile'] * 6
)

# actual minus expected
host_teams['outperformance'] = (
    host_teams['performance_score']
    -
    host_teams['expected_score']
)
```

```python
# hosts only 
hosts_perf = host_teams[
    host_teams['is_host']
]

hosts_perf[
    ['wc_year',
     'team',
     'outperformance']
].sort_values(
    'outperformance',
    ascending=False
)
```

```python
# Statiscal test

host_scores = host_teams[
    host_teams['is_host']
]['outperformance']

non_host_scores = host_teams[
    ~host_teams['is_host']
]['outperformance']

stats.ttest_ind(
    host_scores,
    non_host_scores,
    equal_var=False
)
```

```python
summary = pd.DataFrame({
    "Metric":[
        "Average Goals/Match",
        "Most Successful Confederation",
        "Average Host Outperformance"
    ],
    "Value":[
        tournaments['goals_per_match'].mean(),
        champions['confederation'].value_counts().idxmax(),
        hosts_perf['outperformance'].mean()
    ]
})

summary
```
